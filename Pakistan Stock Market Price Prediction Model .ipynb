# Model Building

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score 
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
X = stock_df[['Open', 'High', 'Low', 'Volume']]  
y = stock_df['Close']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
xscaler = MinMaxScaler(feature_range=(0,1))
yscaler = MinMaxScaler(feature_range=(0,1))
xscaled_data = xscaler.fit_transform(X.values)
yscaled_data = yscaler.fit_transform(y.values.reshape(-1, 1))
yscaled_data
```

```python
def cost_func(x, y, w, b):
    m = x.shape[0]
    total_cost = 0.0
    for i in range(m):
        yhat = np.dot(x[i], w) + b
        total_cost = total_cost + (yhat - y[i]) ** 2
    total_cost = total_cost / (2 * m)
    return total_cost
```

```python
w = [0.0, 0.0, 0.0, 0.0]
b = 0
cost = cost_func(xscaled_data, yscaled_data, w, b)
print(f'Cost at optimal w : {float(cost)}')
```

```python
def compute_gradient(x, y, w, b):
    
    m,n = x.shape           #(number of examples, number of features)
    dj_dw = np.zeros((n,))
    dj_db = 0.0

    for i in range(m):                             
        err = (np.dot(x[i], w) + b) - y[i]
        for j in range(n):                         
            dj_dw[j] = dj_dw[j] + err * x[i, j]    
        dj_db = dj_db + err                        
    dj_dw = dj_dw / m                                
    dj_db = dj_db / m                                
        
    return dj_dw, float(dj_db)
```

```python
df_dw, df_db = compute_gradient(xscaled_data, yscaled_data, w, b)
print(f"Derivative of df_dw: {df_dw}")
print(f"Derivative of df_db: {df_db}")
```

```python
def gradient_descent(x, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters): 
    
    # number of training examples
    m = len(x)
    
    # An array to store cost J and w's at each iteration
    J_history = []
    w_history = []
    w = copy.deepcopy(w_in)  
    b = b_in
    
    for i in range(num_iters):

        # Calculate the gradient and update the parameters
        dj_dw, dj_db = gradient_function(x, y, w, b)  

        # Update Parameters using w, b, alpha and gradient
        w = w - alpha * dj_dw               
        b = b - alpha * dj_db               

        # Save cost J at each iteration
        if i<100:      
            cost =  cost_function(x, y, w, b)
            J_history.append(cost)

        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters/10) == 0:
            w_history.append(w)
            print(f"Iteration {i:4}: Cost {float(J_history[-1]):8.2f}   ")
        
    return w, b, J_history, w_history  #return w and J,w history for graphing
```

```python
# initialize fitting parameters. 
initial_w = np.zeros_like(w)
initial_b = 0.


iterations = 100
alpha = 0.01

w_final, b_final, J_hist, w_history = gradient_descent(xscaled_data, yscaled_data, initial_w, initial_b, cost_func,
                                            compute_gradient, alpha, iterations)
m = xscaled_data.shape[0]
print(f"b,w found by gradient descent: {b_final},{w_final} ")
```

```python
pred = np.dot(xscaled_data, w_final) + b_final
m = xscaled_data.shape[0]
for i in range(m):
    print(f"prediction: {pred[i]:0.2f}, target value: {yscaled_data[i]}")
```

```python
mse = pred - yscaled_data
np.mean(mse)
```

```python
import numpy as np

# Obtain Predictions
y_pred = np.dot(xscaled_data, w_final) + b_final  # Compute predictions using final weights and bias

# Calculate Residuals
residuals = y_pred - yscaled_data  # Subtract actual values from predictions

# Compute MSE
mse = np.mean(residuals ** 2)  # Square the residuals, take the mean

# Print MSE
print(f"Mean Squared Error: {mse}")
```
